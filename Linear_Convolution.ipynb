```python
# import necessary library
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

##Compute linear convolution of two signals and also compare the result with the output of the Matlab/Python library function.##

```python
# user defined convolve function
# use for convolve two signal
# return the convolved signal
def convolve(x_n, h_n):
  y_n = []
  x_len = len(x_n)
  h_len = len(h_n)
  for i in range(x_len + h_len - 1):
    sum = 0
    for k in range(x_len):
      if(i - k >= 0 and i - k < h_len):
        sum += (x_n[k] * h_n[i - k])
    y_n.append(sum)
  return y_n

x_n = [1, 2, 1, -1]
h_n = [1, 2, 3, 1]
y_n_1 = convolve(x_n, h_n)
print("Convolved Signal(With User defined):     ", y_n_1)

y_n_2 = np.convolve(x_n, h_n)
print("Convolved Signal(With library function): ",y_n_2)

plt.figure(figsize=(20,12))
plt.subplot(221)
plt.title("x_n - Input Signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks([i for i in range(len(x_n))])
plt.stem(x_n)

plt.subplot(222)
plt.title("h_n - Unit response signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks([i for i in range(len(h_n))])
plt.stem(h_n)

plt.subplot(223)
plt.title("y_n_1 - Convolved signal(without library function)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks([i for i in range(len(y_n_1))])
plt.stem(y_n_1)

plt.subplot(224)
plt.title("y_n_3 - Convolved signal(with library function)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks([i for i in range(len(y_n_2))])
plt.stem(y_n_2)

plt.show()
```
