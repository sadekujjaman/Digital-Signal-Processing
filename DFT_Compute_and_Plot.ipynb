```python
# import necessary library
import matplotlib.pyplot as plt
import numpy as np
import math
import cmath
%matplotlib inline
```

##Compute and plot Discrete Fourier Transform (DFT) of a signal using DFT equation.##

$$x_k =  \sum_{n=0}^{N-1} x_ne^{\biggl(- \frac{i2\pi}{N}kn\biggl)} $$

```python

def DFT(x_n, N):
  x_k = np.zeros(N,dtype=np.complex)
  for k in range(N):
    temp = 0+0j
    for n in range(N):
      temp += x_n[n]*cmath.exp(-2*cmath.sqrt(-1)*cmath.pi*n*k/N)
    x_k[k] = temp
  
  return x_k

x_n = np.array([4, 7, 3, 2, 3, 1, 4, 5, 9, 6, 4, 2, 1, 3, 5, 6])
x_k = DFT(x_n,len(x_n))

magnitude = np.absolute(x_k)
phase = np.angle(x_k)

plt.figure(figsize=(20,12))
plt.subplot(221)
plt.title("Input Signal For DFT")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.stem(x_n)

plt.subplot(222)
plt.title("Magnitude")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.stem(magnitude)

plt.subplot(223)
plt.title("Phase")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.stem(phase)

plt.show()
```
