```python
# import necessary library
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

#**Plot the elementary signals**#

1. Unit Impulse Signal
2. Unit Step Signal
3. Ramp Signal
4. Exponential Signal


##**Unit Impulse Signal**##

\begin{equation}
  {\delta}[n]=\begin{cases}
    1, & \text{if $n=0$}.\\
    0, & \text{otherwise}.
  \end{cases}
\end{equation}




```python
left = -10
right = 10
n = np.arange(left,right + 1)
delta = np.zeros(n.shape)
delta[np.where(n==0)]=1

plt.figure(figsize=(10,6))
plt.subplot(111)
plt.title("Unit Impulse Signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n, delta)
plt.show()
```

##**Unit Step Signal**##

\begin{equation}
  u[n]=\begin{cases}
    1, & \text{if $n>=0$}.\\
    0, & \text{otherwise}.
  \end{cases}
\end{equation}

```python
left = -10
right = 10
n = np.arange(left,right + 1)
u = np.zeros(n.shape)
u[np.where(n>=0)]=1

plt.figure(figsize=(10,6))
plt.subplot(111)
plt.title("Unit Step Signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n, u)
plt.show()
```

##**Ramp Signal**##

\begin{equation}
  u_r[n]=\begin{cases}
    n, & \text{if $n>=0$}.\\
    0, & \text{otherwise}.
  \end{cases}
\end{equation}

```python
left = -10
right = 10
n = np.arange(left,right + 1)
u_r = np.array([i if i>=0 else 0 for i in n])

plt.figure(figsize=(10,6))
plt.subplot(111)
plt.title("Unit Ramp Signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n, u_r)
plt.show()
```

##**Exponential Signal**##

$$x[n] = a^n,  \text{ for all $n$}$$


```python
left = -10
right = 10
n = np.arange(left,right + 1)
a1 = 0.9    # 0 < a < 1
a2 = 1.1    # a > 1
a3 = -0.9   # -1 < a < 0
a4 = -1.1   # a < -1

x1 = a1**n
x2 = a2**n
x3 = a3**n
x4 = a4**n

plt.figure(figsize=(20,12))
plt.subplot(221)
plt.title("Exponential Signal (0 < a < 1)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n, x1)

plt.subplot(222)
plt.title("Exponential Signal (a > 1)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n, x2)

plt.subplot(223)
plt.title("Exponential Signal (-1 < a < 0)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n, x3)

plt.subplot(224)
plt.title("Exponential Signal (a < -1)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(n)
plt.stem(n, x4)

plt.show()
```
